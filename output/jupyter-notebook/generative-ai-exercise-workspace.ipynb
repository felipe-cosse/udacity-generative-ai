## Step 2 - Prepare an Optional Model Call

The helper below refuses to call a model unless OPENAI_API_KEY is present. It also applies the course's Vocareum endpoint for keys beginning with voc-. Calling a hosted model may use quota, so invoke it deliberately rather than during automatic notebook execution.


```python
def call_model(
    prompt_messages: list[dict[str, str]],
    model: str = "gpt-5-mini",
) -> str:
    """Send prepared messages to a model when credentials are configured."""
    api_key = os.getenv("OPENAI_API_KEY")
    if not api_key:
        return "Skipped: export OPENAI_API_KEY before making a model call."

    litellm.openai_key = api_key
    if api_key.startswith("voc-"):
        litellm.api_base = "https://openai.vocareum.com/v1"

    response = litellm.completion(model=model, messages=prompt_messages)
    return response.choices[0].message.content


print("Model helper ready. Run call_model(messages) only when you want an API call.")
```

## Reference Answer

Run this cell after attempting the exercise. It uses a separate function so your work remains intact.


```python
def factorial_solution(n: int) -> int:
    """Return n factorial for a non-negative integer."""
    if not isinstance(n, int):
        raise TypeError("n must be an integer")
    if n < 0:
        raise ValueError("n must be non-negative")
    if n in (0, 1):
        return 1
    return n * factorial_solution(n - 1)


assert factorial_solution(5) == 120
assert factorial_solution(0) == 1
try:
    factorial_solution(-3)
except ValueError:
    pass
else:
    raise AssertionError("Negative input should raise ValueError")

print("Reference checks passed.")
```
